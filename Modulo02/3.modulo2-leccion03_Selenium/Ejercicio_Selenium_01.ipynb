USAR SELENIUM PARA CREAR UN CÓDIGO QUE HAGA LO SIGUIENTE

1. Escribe un script en Python utilizando Selenium que realice las siguientes tareas en la página web del Decathlon:

2. Abra una el navegador y maximice la ventana.

3. Acceda al sitio web de Decathlon en la URL "https://www.decathlon.es/".

4. Haga clic en el botón "Aceptar" en la notificación de cookies utilizando su selector CSS.

5. Solicite al usuario que ingrese una búsqueda (por ejemplo, un producto deportivo) mediante una entrada de texto.

6. Utilice el texto ingresado para buscar productos en el sitio web de Decathlon y presione la tecla "Enter" para iniciar la búsqueda.

7. Inicialice un diccionario llamado diccionario_resultados con tres listas vacías: "nombre", "precio" y "puntuación".

8. Itere sobre los primeros 5 resultados de la búsqueda y realice las siguientes acciones:

a. Haga clic en un resultado específico. 
b. Recopile el nombre del producto, el precio y la puntuación del producto y almacénelos en las listas correspondientes en el diccionario diccionario_resultados.
c. Navegue de regreso a los resultados de búsqueda.

9. Cierre el navegador Chrome.

El resultado final debería ser un diccionario con información sobre los primeros 5 productos encontrados durante la búsqueda en el sitio web de Decathlon, incluyendo el nombre del producto, el precio y la puntuación. Muestralo en un DataFrame.

```python
# Inicia el navegador Chrome
navegador = webdriver.Chrome()
navegador.get('https://www.decathlon.es/es/')
navegador.maximize_window() # Maximiza la ventana del navegador
sleep(3) # Espera 3 segundos para que la página cargue completamente

# Busca y hace clic en el botón para aceptar las cookies
navegador.find_element('css selector', '#didomi-notice-agree-button > span').click()
print('Ha aceptado las cookies')
sleep(3) # Espera otros 3 segundos

# Escribe 'zapatillas' en la barra de búsqueda y presiona ENTER
navegador.find_element('css selector', '#search-bar > div > form > input').send_keys('zapatillas', Keys.ENTER)
sleep(3) # Espera 3 segundos

# Desplaza la página hacia abajo 1000 píxeles
navegador.execute_script('window.scrollTo(0,1000)')

# Crea un diccionario para almacenar información de productos
info_productos = {'nombre': [], 'precio': [], 'valoracion': []}

# Recorre las primeras 4 páginas de productos
for i in range(1,5):
    sleep(3) # Espera entre cada iteración
    print(f'Va por la iteración {i}')
    navegador.find_element('css selector', f'#app > main > div.list-body.svelte-4gq2qk > section.listing-section.svelte-4gq2qk > div > div:nth-child({i})').click()
    sleep(3) # Espera para que la información del producto cargue

    # Obtiene el nombre, precio y valoración del producto actual y los guarda en el diccionario
    nombre_producto = navegador.find_element("css selector", "#app > main > article > div.vtmn-relative.vtmn-z-\[2\] > section > h1").text
    sleep(2) # Espera antes de obtener el siguiente dato

    precio_producto = navegador.find_element('css selector', '#app > main > article > div.vtmn-relative.vtmn-z-\[2\] > section > div.price-presentation.vtmn-flex.vtmn-flex-col.vtmn-items-start.vtmn-leading-tight.vtmn-mt-2.svelte-1wujduz > div.vtmn-flex.vtmn-flex-col.vtmn-items-start.vtmn-items-stretch.svelte-1wujduz > div > div.vtmn-flex.vtmn-flex-col.vtmn-justify-center.vtmn-items-end > span').text
    sleep(2) # Espera antes de obtener el siguiente dato

    valoracion_producto = navegador.find_element("xpath", '/html/body/div[2]/main/article/div[2]/section/div[3]/div/button/div/span[6]').text

    # Añade los datos al diccionario
    info_productos['nombre'].append(nombre_producto)
    info_productos['precio'].append(precio_producto)
    info_productos['valoracion'].append(valoracion_producto)

    sleep(3) # Espera antes de volver a la página anterior
    navegador.back() # Vuelve a la página anterior

# Cierra el navegador
navegador.close()
print('Ha cerrado la ventana')

# Convierte el diccionario en un DataFrame de pandas
df_productos = pd.DataFrame(info_productos)
df_productos
```
